```python
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import numpy as np





#Normal random noise plus linear relation correlation, a toy model of having a prediction listed for each ticker, where Y is the prediction X is the target
randArrX = np.random.rand(5000)
randArrY = np.random.rand(5000)   + randArrX * -0.13

#Plot target vs prediction assuming that /np.random.rand(5000)   + randArrX * -0.13/ approximates having a dataset that has some linear correlation with the target by combining random noise with the target
plt.scatter(randArrX,randArrY)
print(spearmanr(randArrX,randArrY))
plt.show()

#The same relation between targets and the predictive variable, but predictive data is not available for some tickers, so we fill them with the median of the predictions on available data

#4000 tickers have predictions since data is available
randArrY[1000:] = np.random.rand(4000)   + randArrX[1000:] * -0.13

#1000 tickers get median filled since we have no data to predict them with
randArrY[:1000] = np.median(randArrY[1000:])

#Plot and get correlation for missing submissions example

plt.scatter(randArrX,randArrY)
print(spearmanr(randArrX,randArrY))
```
